#Laughing at Tweets with Poisson Distribution

####Question 5.1

Find $P(3)$: $$P(3) = \frac{e^{-0.5}*0.5^{3}}{3!}$$

```python

N = 500
p = 0.001
mu = N * p
n = 3
P_3 = (exp(-mu)*mu**n) / factorial(n)
print 'Probability you laugh at three tweets is {:.2%}' .format(P_3)
```

####Question 5.2

Find $P(1) + P(2)$: $$P(1) + P(2) = \frac{e^{-0.5}*0.5^{1}}{1!} + \frac{e^{-0.5}*0.5^{2}}{2!}$$

```python
N = 500 
p = 0.001
n_1 = 1
n_2 = 2
mu = N * p
P_less_than_3 = (exp(-mu)*mu**n_1) / factorial(n_1) + (exp(-mu)*mu**n_2) / factorial(n_2)
print 'Probability you laugh less than three times is {:.2%}' .format(P_less_than_3)
```

####Question 5.3

Find $P(n>3)$: $$P(n) = \frac{e^{-\mu}*\mu^{n}}{n!}$$ with n being an array from 4 to 500.

```python
from scipy.misc import comb, factorial
N = 500
p = 0.001
mu = N * p
n = np.arange(4,501)
prob = np.sum((exp(-mu)*mu**n) / factorial(n))
print 'Probabilbility you laugh more than three times is {:.2%}' .format(prob)
```

#Weekend Dating with Binomial Distribution

####Question 6.1

$$Binomial Distribution: Q = [0,N]$$
Find $P(n=3)$: $$P(n) = {N \choose n} * p^{n} * (1 - p)^{N-n}$$

```python
from scipy.misc import comb, factorial
N = 8
n = 3
p = 0.10
prob = comb(N,n) * p**n * (1 - p)**(N-n)
print 'Probability you go on three second dates is {:.2%}' .format(prob)
```

####Question 6.2

Find $P(n>3)$: $$P(n) = {N \choose n} * p^{n} * (1 - p)^{N-n}$$

```python
N = 8
n = np.arange(4,9)
p = 0.10
prob_1 = np.sum(comb(N,n) * p**n * (1 - p)**(N-n))
print 'Probability you go on more than three second dates is {:.2%}' .format(prob_1)
```
